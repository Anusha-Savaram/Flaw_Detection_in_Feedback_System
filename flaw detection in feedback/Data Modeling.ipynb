```python
def filling_bundle_scores_dvalue(cols):
    d_score_1=cols[0]
    d_score_2=cols[1]
    d_score_3=cols[2]
    d_score_4=cols[3]
    d_score_5=cols[4]
    d_score_6=cols[5]
    d_score=cols[6]
    if pd.isnull(d_score):
        return (d_score_1+d_score_2+d_score_3+d_score_4+d_score_5+d_score_6)
    else:
        return d_score
```

```python
def filling_bundle_scores_rvalue(cols):
    r_score_1=cols[0]
    r_score_2=cols[1]
    r_score_3=cols[2]
    r_score_4=cols[3]
    r_score_5=cols[4]
    r_score=cols[5]
    if pd.isnull(r_score):
        return (r_score_1+r_score_2+r_score_3+r_score_4+r_score_5)
    else:
        return r_score
```

```python
def filling_bundle_scores_ovalue(cols):
    o_score_1=cols[0]
    o_score_2=cols[1]
    o_score_3=cols[2]
    o_score_4=cols[3]
    o_score_5=cols[4]
    o_score_6=cols[5]
    o_score=cols[6]
    if pd.isnull(o_score):
        return (o_score_1+o_score_2+o_score_3+o_score_4+o_score_5+o_score_6)
    else:
        return o_score
```

```python
def filling_bundle_scores_svalue(cols):
    s_score_1=cols[0]
    s_score_2=cols[1]
    s_score_3=cols[2]
    s_score_4=cols[3]
    s_score_5=cols[4]
    s_score=cols[5]
    if pd.isnull(s_score):
        return (s_score_1+s_score_2+s_score_3+s_score_4+s_score_5)
    else:
        return s_score
```

```python
def filling_bundle_scores_uvalue(cols):
    u_score_1=cols[0]
    u_score_2=cols[1]
    u_score_3=cols[2]
    u_score_4=cols[3]
    u_score_5=cols[4]
    u_score_6=cols[5]
    u_score=cols[6]
    if pd.isnull(u_score):
        return (u_score_1+u_score_2+u_score_3+u_score_4+u_score_5+u_score_6)
    else:
        return u_score
```

```python
def filling_bundle_scores_cesd_score(cols):
    c_score_1=cols[0]
    c_score_2=cols[1]
    c_score_3=cols[2]
    c_score_4=cols[3]
    c_score_5=cols[4]
    c_score_6=cols[5]
    c_score_7=cols[6]
    c_score_8=cols[7]
    c_score_9=cols[8]
    c_score_10=cols[9]
    cesd_score=cols[10]
    if pd.isnull(cesd_score):
        return (c_score_1+c_score_2+c_score_3+c_score_4+c_score_5+c_score_6+c_score_7+c_score_8+c_score_9+c_score_10)
    else:
        return cesd_score
```
